        // 获取html的标签
        var tag1 = document.getElementById("text1");
        // 获取标签的文字
        var text1 = tag1.innerText;
        // 拼接新字符串

        var newStr =  otherStr + first_char;

## DOM操作
DOM:Document Object Model,网页结构的对象模型。
- 获取元素

- 修改元素内容
element.innerText="new content";
- 修改元素样式
element.style.color="red";
- 添加子元素
parentElement.appendChild(childElement);
- 删除子元素
parentElement.removeChild(childElement);
- 获取父元素
parentElement=childElement.parentNode;
- 获取所有子元素
childrenElements=parentElement.children;
- 创建元素
document.createElement("tagName");

## BOM操作
BOM:Browser Object Model,浏览器对象模型。
- window对象是浏览器窗口的全局对象，可以访问到整个浏览器窗口和页面的所有元素。
window.location.href; //当前页面的URL地址
window.screen.width; //屏幕宽度
window.screen.height; //屏幕高度
window.screen.availHeight; //可视区域的高度(不包括工具栏)
